```python
# Import libraries needed

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data

df = pd.read_csv(r'/Users/amei/Downloads/movies.csv')
```

```python
# Look at the data

df.head()
```

```python
# Remove any missing data

df = df.dropna()

# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for columns

df.dtypes
```

```python
# Update budget, gross, and votes format

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
# Create correct year column by taking year of released

df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Budget high correlation ?
# Company high corelation ?

# Scatter plot with budget vs gross 

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df.head()
```

```python
# Use Seaborn to plot the Budget vs Gross

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Look at correlation

df.corr(method='pearson') # different methods: pearson, kendall, spearman
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)


plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at Company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)


plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

pd.set_option('display.max_rows', None)

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs 
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr 
```
